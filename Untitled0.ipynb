```python
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
import seaborn as sns
appId="285e2461d257bea5d0152297fbb32b7c83727aad"
url="http://api.e-stat.go.jp/rest/3.0/app/json/getStatsData?"
param = {
    "appId"            : appId,
    "lang"             : "J",
    "statsDataId"      : "0003240561",
    "metaGetFlg"       : "Y",
    "cntGetFlg"        : "N",
    "explanationGetFlg" : "Y",
    "annotationGetFlg" : "Y",
    "sectionHeaderFlg" : "1",
    "replaceSpChars" : 0
}

response=requests.get(url,params=param)
data=json.loads(response.text)
json=json.dumps(data,indent= 2, ensure_ascii=False)
#print(response.url)
#print(json)
values = data['GET_STATS_DATA']['STATISTICAL_DATA']['DATA_INF']['VALUE']
print(values)

x_l=[]
x_s=[]
y_l=[]
y_s=[]

#cat01学部(100:全体, 110:文系, 120:理系) cat02が学校区分(140：国立, 150:私立)
for value in values: 

  if value["@cat01"] == "110"  and value["@cat02"]=="130" and value["@tab"] == "120": 
    x_l.append(int(value["@time"][:4])) #年度を追加
    y_l.append(float(value["$"])) #内定率を追加

  elif value["@cat01"] == "120" and value["@cat02"]=="130" and value["@tab"] == "120":
    x_s.append(int(value["@time"][:4]))
    y_s.append(float(value["$"]))

print(y_l)

plt.figure(figsize=(15,5))#表示サイズの設定
plt.plot(x_l, y_l, color="b", label="l_job_offer_rate") 
plt.plot(x_s, y_s, color="r", label="s_job_offer_rate") 
plt.legend(loc="lower left", fontsize=15) #凡例表示

plt.show()


for value in values:
   value["@time"]=value["@time"][:4]

df=pd.DataFrame(values)
print(df)
print(df.dtypes)
df["@time"]=df["@time"].astype(int)

l_df=df[(df["@cat01"]=="110") & (df["@cat02"]=="130") & (df["@tab"]=="120")]
s_df=df[(df["@cat01"]=="120") & (df["@cat02"]=="130") & (df["@tab"]=="120")]
print(l_df)
l_df["$"]=l_df["$"].astype(float)
s_df["$"]=s_df["$"].astype(float)

plt.figure(figsize=(15,5))
plt.plot(l_df["@time"],l_df["$"],color="b")
plt.plot(s_df["@time"],s_df["$"],color="r")
plt.show()
```
